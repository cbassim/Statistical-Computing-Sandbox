```python
%reset
# Importing Necessary Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Jupyter Notebook Example: ScatterPlots_WineQuality.ipynb

```python
# load in the Wine Quality csv
Wine=pd.read_csv("WineQuality.csv")
```

```python
Wine
```

```python
# let's go ahead and work with seaborn
# lets set the theme we want seaborn to use
sns.set_theme(style='ticks')
```

```python
sns.scatterplot(x="Color", y="Quality", data=Wine)
```

```python
# pairplot() plot pairwise relationships:  Scatterplot Matrix
sns.pairplot(Wine)
```

```python
from scipy import stats
stats.pearsonr(x=Wine["Color"],y=Wine["Quality"])
```

```python
correlation_matrix=Wine.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,cmap="coolwarm")
```

```python
# Monty equation for correlation
np.sum( ( Wine["Quality"]-np.mean(Wine["Quality"]))  * ( Wine["Color"]-np.mean(Wine["Color"])) )
```

```python
np.sum( ( Wine["Quality"])  * ( Wine["Color"]-np.mean(Wine["Color"])) )
```
