# Pie Charts and Bar Charts and Category Charts

```python
Tx = pd.read_csv('Comparing2Treatments.csv')
```

```python
Tx
```

```python
Tx=Tx.sort_values(by=["Coating","SurfRough"])
```

```python
%matplotlib inline
plt.clf()
plt.close()

Pie1=Tx[Tx["Coating"]=="A"]
Pie2=Tx[Tx["Coating"]=="B"]
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(Pie1['SurfRough'],labels=Pie1['ProductSeries'])
ax2.pie(Pie2['SurfRough'],labels=Pie2['ProductSeries'])
```

```python
g=sns.barplot(x=Tx['Coating'],y=Tx['SurfRough'],hue=Tx['ProductSeries'])
sns.move_legend(g, "upper left")
```

```python
sns.catplot(x=Tx['Coating'],y=Tx['SurfRough'],hue=Tx['ProductSeries'])
sns.lineplot(x=Tx['Coating'],y=Tx['SurfRough'],hue=Tx['ProductSeries'],legend=False)
```
