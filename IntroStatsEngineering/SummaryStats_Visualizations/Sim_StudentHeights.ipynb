# Probability Plot

```python
from scipy import stats
# Requires speciallty lined probability paper for the normal distribiution
import probscale
```

```python
fig2, (ax2) = plt.subplots(1)
fig2=probscale.probplot(SampleStudentHeights['Height'],ax2)
```

```python
#Generates a probability plot of sample data against the quantiles of a specified theoretical distribution 
fig,ax = plt.subplots()
stats.probplot(SampleStudentHeights['Height'],dist='norm',plot=ax)
plt.show()
```
